#Random Forests

#1 Booststrap sampling the training sata

กำหนดให้ N มีข้อมูลอยู่ 100 ตัว จากนั้นแบ่งตัวแปร train และ test

```
N = 100 # number of trees in the forest
```

```
training_data = list()
for i in range(N): ##ต้นไม้แต่ละอันเห็น data = 80%
  x_sp, x_dp, y_sp, y_dp = train_test_split(X_trian, y_train, test_size=0.2, random_state = i+6)
  training_data.append({'X':x_sp,'y':y_sp})
```

#2 Train separately

ทำการTrain ข้อมูลโดยการวนลูปแล้วเก็บไว้ในตัวแปร classifier_forest

```
classifier_forest = list()
for i in range(N):
  tree = DecisionTreeClassifier(max_features='sqrt',random_state=i+2)
  tree.fit(training_data[i]['X'], training_data[i]['y'])
  classifier_forest.append(tree)
```

#3 To predict,aggregating

ทำการทำนายค่าใน classifier_forest ในทุกๆ i แล้วเก็บค่าตัวแปรไว้ใน prediction

```
prediction = list()
for i in range(N):
  pred = classifier_forest[i].predict(X_test)
  #print(pred)
  prediction.append(pred)
```

```
prediction
```

เรียกใช้คำสั่ง stats เพื่อใช้วิธีการทางสถิติ

```
from scipy import stats
import numpy as np
```

ให้ตัวแปร m เก็บค่าฐานนิยมของตัวแปร prediction

```
m = stats.mode(np.array(prediction))
```

```
m
```

#4 Evaluation

หาค่าAccuracy

```
from sklearn.metrics import accuracy_score
```

```
accuracy_score(y_test, m.mode.squeeze())
```
